### ETH-USD

```python
# 1. Install yfinance if you haven't
# pip install yfinance

import yfinance as yf
import pandas as pd

# 2. Define ticker, start, and end dates
ticker = "ETH-USD"
start_date = "2023-11-23"
end_date = "2025-12-05"

# 3. Download data
eth_data = yf.download(ticker, start=start_date, end=end_date)

# 4. Reset index to have 'Date' as a column
eth_data = eth_data.reset_index()

# 5. Save to CSV
eth_data.to_csv("eth_usd_history.csv", index=False)

print("Data downloaded successfully!")
print(eth_data.head())
```

```python
# Install yfinance if not installed
# pip install yfinance

import yfinance as yf
import pandas as pd

# List of crypto tickers on Yahoo Finance
cryptos = {
    "Bitcoin": "BTC-USD",
    "Ethereum": "ETH-USD",
    "Tether": "USDT-USD",
    "Binance Coin": "BNB-USD",
    "Cardano": "ADA-USD",
    "Solana": "SOL-USD",
    "Ripple": "XRP-USD",
    "Polkadot": "DOT-USD",
    "Avalanche": "AVAX-USD",
    "Polygon": "MATIC-USD"
}

start_date = "2023-11-23"
end_date = "2025-12-10"

for name, ticker in cryptos.items():
    print(f"Downloading {name} ({ticker}) data...")
    data = yf.download(ticker, start=start_date, end=end_date)
    data = data.reset_index()  # Make 'Date' a column
    filename = f"{ticker}_daily_history.csv"
    data.to_csv(filename, index=False)
    print(f"Saved {filename}, {len(data)} rows.\n")

print("All crypto data downloaded successfully!")
```
