```python
import pandas as pd

# 1. LOAD ETH PRICE DATA
eth_path = r"C:\Users\moham\Downloads\scrape\main\crypto_project\data\raw\eth_usd_daily.csv"

# Skip the second row which is repeated headers
eth_df = pd.read_csv(eth_path, skiprows=[1])  # skip row index 1

print("Original ETH columns:", eth_df.columns)

# 2. PARSE DATE (dayfirst=True because format is DD-MM-YYYY)
eth_df["Date"] = pd.to_datetime(eth_df["Date"], dayfirst=True, errors='coerce')
eth_df = eth_df.dropna(subset=["Date"])  # keep valid dates only

# 3. KEEP ONLY USEFUL COLUMNS
eth_df = eth_df[["Date", "Open", "High", "Low", "Close", "Volume"]]

# 4. RENAME COLUMNS
eth_df = eth_df.rename(columns={
    "Date": "date",
    "Open": "eth_open",
    "High": "eth_high",
    "Low": "eth_low",
    "Close": "eth_close",
    "Volume": "eth_volume"
})

# 5. SORT (IMPORTANT FOR MERGING)
eth_df = eth_df.sort_values("date").reset_index(drop=True)

print("Prepared ETH dataset:")
print(eth_df.head())


# 6. OPTIONAL SAVE
eth_df.to_csv(r"C:\Users\moham\Downloads\scrape\main\crypto_project\data\processed\eth_cleaned_yf.csv", index=False)
```
